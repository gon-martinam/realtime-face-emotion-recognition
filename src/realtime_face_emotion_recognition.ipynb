## Imports

```python
import tensorflow as tf
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_directory = './drive/MyDrive/data_face_recognition/train/'
classes = ['0', '1', '2', '3', '4', '5', '6']
```

```python
img_size = 224 # ImageNet -> 224 x 224
train_data_list = []

def create_train_data():
    for category in classes:
        path = os.path.join(data_directory, category)
        class_num = classes.index(category)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path, img))
                new_array = cv2.resize(img_array, (img_size, img_size))
                train_data_list.append([new_array, class_num])
            except Exception as e:
                pass
```

```python
create_train_data()
```

```python
print(len(train_data_list))
```

```python
import random

random.shuffle(train_data_list)
```

```python
X = []
y = []

for features,label in train_data_list:
  X.append(features)
  y.append(label)

X = np.array(X).reshape(-1, img_size, img_size, 3) ## converting it to 4 dimensions
```
